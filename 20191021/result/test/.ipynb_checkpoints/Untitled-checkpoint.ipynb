```python
import xlwt, xlrd
wb = xlrd.open_workbook('posts.xlsx', encoding_override='gbk')
sheet = wb.sheet_by_index(0)
all_post = []
for i in range(sheet.nrows):
    idx, p_time, text, url = sheet.row_values(i)
    all_post.append((int(idx), p_time, text, url))
print(len(all_post))
```

```python
from bs4 import BeautifulSoup
import re

re_commentlink      = '<a(?:.[^>]*?)href="(.[^"]*?)"(?:[^>]*?)>(?:[0-9]* Comments|1 Comment|Comment)</a>'
pattern_commentlink = re.compile(re_commentlink)

with open('posts.html', 'rb') as f:
    a = f.read()
soup = BeautifulSoup(a.decode('utf-8'), 'lxml')
```

```python
from urllib.parse import unquote, quote # url编码解码

def url_decode(s: str):
    return unquote(s).replace('amp;', '')

divs = soup.find_all('div', attrs={'data-sigil': "reactions-bling-bar"})
articles = soup.find_all('article')
for idx, item in enumerate(articles):
    footer = item.find('footer')
    if footer:
        cnt_link = str(pattern_commentlink.findall(str(item))[0])
        content = footer.find('div', attrs={'data-sigil':"reactions-bling-bar"})
#         print("-"*30)
#         print(cnt_link)
        num_reaction = num_share = num_comment = 0
        if content:
            reaction = content.find('div', attrs={"data-sigil":"reactions-sentence-container"})
            if reaction:
                num_reaction = str(reaction.text)
                num_reaction = re.sub('\s+', ' ', num_reaction)
                info = num_reaction.split()[1:]
                if len(info) > 1:
                    num_reaction = info[2]
                elif len(info) == 1:
                    num_reaction = info[0]
                else:
                    num_reaction = 0
            comment = re.findall(r'([0-9]+)\s[Comment|Comments]', str(content))
            if comment:
                num_comment = comment[0]
            share = re.findall(r'([0-9]+)\s[Share|Shares]', str(content))
            if share:
                num_share = share[0]
        link = 'https://m.facebook.com' + url_decode(cnt_link)
        print(link[0:81], num_reaction, num_comment, num_share)
        for i in range(len(all_post)):
            if all_post[i][3].startswith(link[0:81]):
                all_post[i] = all_post[i] + (num_reaction, num_comment, num_share)
```

```python
wb = xlwt.Workbook()
sheet = wb.add_sheet('posts')
for idx, value in enumerate(all_post):
    for cnt, val in enumerate(value):
        print(idx, val)
        sheet.write(idx, cnt, val)
wb.save('posts_adddatas.xls')
```
